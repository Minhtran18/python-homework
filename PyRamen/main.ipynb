```python
# Import from the pathlib library
from pathlib import Path
import csv

# Check the current directory where the Python program is executing from
print(f"Current Working Directory: {Path.cwd()}")

#set path file
menu_filepath = Path("02-Python_PyRamen_Resources_menu_data.csv")
sales_filepath = Path("02-Python_PyRamen_Resources_sales_data.csv")

#initialize empty menu(list), sales(list), report(dict):
menu = []
sales = []
row_count = 0
#Initialize dict object to hold our key-value pairs of items and metrics
report = {}


#open and read menu_data:
with open(menu_filepath, 'r') as csvfile:
    
    csvreader = csv.reader(csvfile, delimiter=',')
    #skip header:
    header =next(csvreader)
 
    #Loop over the rest of the rows 
    for row in csvreader:
        #append every row to the menu lit
        menu.append(row)

#open and read sales_data:
with open(sales_filepath, 'r') as csvfile:
    
    csvreader = csv.reader(csvfile, delimiter=',')
    #skip header:
    header =next(csvreader)
 
    #Loop over the rest of the rows 
    for row in csvreader:
        #append every row to the sales list
        sales.append(row)
    
    
#loop in sales list object:
for salerow in sales:
    
    #set the following columns of the sales data to their own variables:
    Quantity = salerow[3]
    sales_item = salerow[4]
   
    #Perform a quick check if the sales_item is already included in the report.
    # If the item value not in the report, add it as a new entry with initialized metrics
    if sales_item not in report:
        #Create a nested loop by looping through every record:
        report[sales_item] = {"01-count": 0,"02-revenue": 0,"03-cogs": 0,"04-profit": 0}
    else:
        continue

#For every row in our sales data, loop over the menu records to determine a match:
for salerow in sales:
    #set the following columns of the sales data to their own variables:
    Quantity = salerow[3]
    sales_item = salerow[4]
          
    #loop through every row in the menu list object:
    for menurow in menu:
        #Initialize menu data variables:
        menu_item = menurow[0]
        menu_price = menurow[3]
        menu_cost = menurow[4]
        row_count+=1
        #If sales_item equal item in menu, use quantity in sales_data and the price & cost in menu_data to calculate profit for each item
        if sales_item == menu_item:
            report[sales_item]["01-count"] += int(Quantity)
            report[sales_item]["02-revenue"] += (float(menu_price) * int(Quantity))
            report[sales_item]["03-cogs"] += (int(menu_cost) * int(Quantity))

            # @TODO: Else, the sales item does not equal any fo the item in the menu data, therefore no match
        else:
            continue
            
            
for item, valuedict in report.items():
    #loop key in dict 
        for key in valuedict:
        #set variables COGS and REVENUE to their own key in report {dict}:
            COGS = report[item]["03-cogs"]
            revenue = report[item]["02-revenue"]
        #key equal profit:
            if key == "04-profit":
        #formula to get profit:
                report[item][key] = revenue - COGS

#output file path and name of text file
output_path = Path("PyRamen.txt")
# Open the output_path as a file object in "write" mode ('w')
with open(output_path, 'w') as file:
    # Write a header line and write the contents of 'text' to the file:
    file.write("This is the financial report for PyRamen.\n")
    for key in report:
        #write w end of line:
        file.write(f"{key} {report[key]} \n")   
```
