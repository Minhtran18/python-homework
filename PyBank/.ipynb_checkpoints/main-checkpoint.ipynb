```python
#import file:
from pathlib import Path
import csv
#print(f"Current Working Directory: {Path.cwd()}")

#set path file
csvpath = Path("PyBank_budget_data.csv")

#Initializ variables
total_months = 0
total_profit_loss = 0
previous_profit_loss = 0
profit_loss_change = 0

greatest_increase = ["", 0]
greatest_decrease = ["", 9999999]


# Open the file in "read" mode ('r') and store the
with open(csvpath, 'r') as csvfile:
    
    csvreader = csv.reader(csvfile, delimiter=',')
    #skip header:
    header =next(csvreader)
 
    #loop get total months and PnL:
    for row in csvreader:
        
        #total_months = total_months +1
        total_months += 1
        
        #The net total amount of Profit/Losses over the entire period.
        total_profit_loss = total_profit_loss + int(row[1])
        
        #The average of the changes in Profit/Losses over the entire period.
        average_profit_losses = total_profit_loss/int(total_months)
        
        #list profit_loss_change:
        profit_loss_change = int(row[1]) - previous_profit_loss
        
        #get previous_profit_loss:
        previous_profit_loss = int(row[1])
      
              
       #if function to find greatest increase and decrease:
        if (profit_loss_change > greatest_increase[1]):
            greatest_increase[0] = row[0]
            greatest_increase[1] = profit_loss_change
            
        if (profit_loss_change < greatest_decrease[1]):
            greatest_decrease[1] = profit_loss_change
            greatest_decrease[0] = row[0]
            
# show final result:
print("Financial Analysis")
print("-------------------------")
print("Total Months: " + str(total_months))
print("Total : " + "$" + str(total_profit_loss))    
print("Average Change: " + "$" + str(round(average_profit_losses,2)))
print("Greatest Increase: " + str(greatest_increase[0]) + " ($" +  str(greatest_increase[1]) + ")") 
print("Greatest Decrease: " + str(greatest_decrease[0]) + " ($" +  str(greatest_decrease[1]) + ")")
            

# Set the output file path
output_path = Path("output.txt")
# Open the output_path as a file object in "write" mode ('w')
# extra line break to each line in the output.
with open(output_path, 'w') as file:
      
    #Writing output to the file
    file.write("This is an output file.\n")
    file.write("Financial Analysis \n")
    file.write("------------------------- \n")
    file.write("Total Months: " + str(total_months) + "\n")
    file.write("Total : " + "$" + str(total_profit_loss) + "\n")    
    file.write("Average Change: " + "$" + str(round(average_profit_losses,2)) + "\n")
    file.write("Greatest Increase: " + str(greatest_increase[0]) + " ($" +  str(greatest_increase[1]) + ")\n") 
    file.write("Greatest Decrease: " + str(greatest_decrease[0]) + " ($" +  str(greatest_decrease[1]) + ")\n")

```
